# Job Seekers Data

```python
df1 = pd.read_csv("data/jobseekers.csv")
```

```python
df1short_names = [
    'Occupation',
    'AgencyComments',
    'Age',
    'EducationLevel',
    'UsedAgency',
    'LastUsedAgency',
    'TimeToSecureJob',
    'SecuredJob',
    'SatisfiedServices',
    'Challenges',
    'AgencyImprovements',
    'LikelihoodToRecommend',
    'Gender',
    'Satisfaction',
    'OverallExperience',
    'JobSearchMethods',
    'MostUtilizedMethod',
    'AgencyImportance',
    'EmploymentStatus',
    'LeastSatisfiedServices'
]
```

```python
df1.columns = df1short_names
```

```python
#remove white space after word
df1 = df1.apply(lambda x: x.str.strip() if x.dtype=="object" else x)
```

```python
df1.head()
```

```python
df1['Occupation'].info
```

```python
plt.figure(figsize=(12,4), dpi=150)
sns.countplot(data=df1, x='Gender')
```

```python
plt.figure(figsize=(12,4), dpi=150)
sns.countplot(data=df1, x='JobSearchMethods')
plt.xticks(rotation=90)
```

# Recruitment Agency Data

```python
df3 = pd.read_csv('recruitment_agency.csv')
```

```python
df3 = df3.drop(["Timestamp","Unnamed: 17", "Unnamed: 18", "Unnamed: 19", "Unnamed: 20", "Unnamed: 21", "Email address"], axis=1)
```

Org Name:

Explanation: The name of the organization.
Org Email:

Explanation: The email address of the organization.
Phone:

Explanation: The phone number of the organization.
Years in Op:

Explanation: The number of years the organization has been in operation.
Operational Regions:

Explanation: The regions in Ghana where the organization primarily operates.
Competition Level:

Explanation: A description of the level of competition in the organization's operational region(s).
Services Offered:

Explanation: The services offered by the organization. (Multiple options can be selected)
Most In-Demand Service:

Explanation: The service among the offerings that the organization believes is most in-demand.
Client Duration:

Explanation: The typical duration for which clients work with the organization.
Client Industries:

Explanation: The industries from which the majority of the organization's clients come.
Market Trend (5 Years):

Explanation: The market trend over the last five years.
Key Services:

Explanation: The services for which the organization is best known.
Primary Challenges (Select 2):

Explanation: The primary challenges the organization faces in the recruitment market in Ghana. (Select up to 2)
Emerging Trends/Opportunities:

Explanation: Any emerging trends or opportunities the organization sees in the Ghanaian recruitment market.
Unmentioned Services/Trends:

Explanation: Any services or trends not mentioned in the questionnaire that the organization believes are important in the Ghanaian recruitment landscape.

```python
newcolumn_names = [
    "Org Name", "Org Email", "Phone", "Years in Op", "Operational Regions",
    "Competition Level", "Services Offered", "Most In-Demand Service",
    "Client Duration", "Client Industries", "Market Trend (5 Years)",
    "Key Services", "Primary Challenges (Select 2)",
    "Emerging Trends/Opportunities", "Unmentioned Services/Trends"
]

# Rename the DataFrame columns
df3.columns = newcolumn_names
```

```python
df3
```

```python
df3.info()
```

```python
plt.figure(figsize=(12,4), dpi=150)
sns.countplot(data=df3, x='Operational Regions')
```

```python
plt.figure(figsize=(12,4), dpi=150)
sns.countplot(data=df3, x='Most In-Demand Service', palette='magma')
plt.xticks(rotation=30)
```

```python
plt.figure(figsize=(12,4), dpi=150)
sns.countplot(data=df3, x='Competition Level', palette='magma')
plt.xticks(rotation=30)
```
