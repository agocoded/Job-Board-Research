# Load the data

```python
dfu = pd.read_csv("updated_file.csv")
```

```python
dfu
```

# Data Cleaning

**Droping the Timestamp Column**

```python
dfu = dfu.drop("TIMESTAMP", axis=1)
```

```python
#Hadling excess spaces for strings
dfu = dfu.apply(lambda x: x.str.strip() if x.dtype=="object" else x)
```

```python
#Replace instances of S.H.S to SHS on "EDUCATIONAL LEVEL" feature
dfu['EDUCATION LEVEL'] = dfu['EDUCATION LEVEL'].str.replace('S.H.S','SHS')
```

# Visualizations

```python
plt.figure(figsize=(12,4),dpi=150)
sns.countplot(data=dfu,x='GENDER',)
```

```python
plt.figure(figsize=(12,4),dpi=150)
sns.countplot(data=dfu,x='AGE RANGE',palette='magma')

plt.title("Visualization of Age Range")
```

**Job Board Sources**

```python
plt.figure(figsize=(12,4),dpi=150)
sns.countplot(data=dfu,x='OTHER JOB SEARCH SOURCES',palette='magma')
plt.xticks(rotation=90)
plt.title("Visualization of Job Board Sources")
```

**Educational Levels**

```python
plt.figure(figsize=(12, 4), dpi=150)
ax = sns.countplot(data=dfu, x='EDUCATION LEVEL')
plt.xticks(rotation=90)
```

# Board Ratings

```python
# Group the data by 'USED JOB BOARDS' and calculate the average rating
average_ratings = dfu.groupby('USED JOB BOARDS')['JOB BOARD SUCCESS RATING', 'JOB BOARD EFFICIENCY RATING', 'JOB POSTING QUALITY RATING', 'LIKELIHOOD TO RECOMMEND JOB BOARDS'].mean().reset_index()

average_ratings
```

```python
# Create a bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=average_ratings, x='USED JOB BOARDS', y='JOB BOARD SUCCESS RATING', palette='viridis')
plt.title('Average Job Board Success Rating')
plt.xticks(rotation=90)  # Rotate x-labels for readability
```

```python
# Create a bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=average_ratings, x='USED JOB BOARDS', y='JOB BOARD EFFICIENCY RATING', palette='viridis')
plt.title('Average Job Board Efficiency Rating')
plt.xticks(rotation=90)  # Rotate x-labels for readability
```

```python
# Create a bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=average_ratings, x='USED JOB BOARDS', y='JOB POSTING QUALITY RATING', palette='viridis')
plt.title('Average Job Board Quality Rating')
plt.xticks(rotation=90)  # Rotate x-labels for readability
```

```python
# Create a bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=average_ratings, x='USED JOB BOARDS', y='LIKELIHOOD TO RECOMMEND JOB BOARDS', palette='viridis')
plt.title('Average Job Board Recommedation Rating')
plt.xticks(rotation=90)  # Rotate x-labels for readability

```
